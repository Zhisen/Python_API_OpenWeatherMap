```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint
# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# import api_key from the os
import os
api_key = os.environ["OpenWeatherMap_API"]
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# Generate 2500 sets of coordinates
np.random.seed(0)
x = np.random.uniform(low = -90,high = 91,size = 2500)
y = np.random.uniform(low = -180,high = 181,size = 2500)
coordinates = list(zip(x,y))
city_country_list = []
lats = []
found_cities = []
temperatures = []
humidity = []
cloudiness = []
wind_speed = []
counter = 0

# Find the nearest cities, and create a list of unique (non-repeated) cities
for item in coordinates:
    nearest_city = citipy.nearest_city(*item)
    city_country = f'{nearest_city.city_name}, {nearest_city.country_code}'
    if city_country not in city_country_list:
        city_country_list.append(city_country)

# Request info from the OpenWeatherMap api for each city that the website has info for
for city in city_country_list:
    try:
        response = requests.get(url + '&q=' + city).json()
        lat = response['coord']['lat']
        temp = response['main']['temp']
        humi = response['main']['humidity']
        cloudi = response['clouds']['all']
        wind = response['wind']['speed']
        lats.append(lat)
        temperatures.append(temp)
        humidity.append(humi)
        cloudiness.append(cloudi)
        wind_speed.append(wind)
        found_cities.append(city)
        counter += 1
        print(f'{counter}: {city.upper()} has a temperture at {temp} degrees F, humidity at {humi}%, cloudiness at {cloudi}%, and wind speed at {wind} mph.')
    except KeyError:
        print("---- Oh uh, city not found, but don't worry, we still have more than 500 cities")
```

```python
length = len(found_cities)
print(f'There are a total of {length} cities found.')
```

```python
#save as a datafrme, and then onto a csv file
results = pd.DataFrame({
    'Number': np.arange(1,length+1,1),
    'City,CountryCode': found_cities,
    'Temperature (F)': temperatures,
    'Humidity (%)': humidity,
    'Cloudiness (%)': cloudiness,
    'Wind Speed (mph)': wind_speed,
})
results.to_csv("results.csv")
results.head()
```
